## creation of the data

```python
import numpy as np
np.random.seed(42)
window = 3

X = np.zeros((len(mapped_text) - 2 * window, 2 * window))
Y = np.zeros((len(mapped_text) - 2 * window))
```

```python
for i in range(window, len(mapped_text) - window):
    Y[i-window] = mapped_text[i]
    X[i-window] = mapped_text[i-window:i+window]
```

```python
X = X.astype(int)
Y = Y.astype(int)
```

## Embeddings model

```python
def softmax(X):
    return np.exp(X) / np.sum(np.exp(X))

def crossEntropy(y, y_pred):
    return -np.log(y @ y_pred) 

class embedder:
    def __init__(self, window, voc_size, K = 20, cnn = False):
        self.voc_size = voc_size
        self.window = window
        self.K = K
        self.cnn = cnn
        self.W_I = np.random.randn(voc_size, K) / np.sqrt(K)
        self.W_O = np.random.randn(K, voc_size) / np.sqrt(K)
        
    def to_1_hot(self, X, isY = False):
        if isY:
            res = np.zeros(self.voc_size)
            res[X] = 1
        else:
            res = np.zeros((self.voc_size, len(X)))
            for (i, x) in enumerate(X):
                res[x][i] = 1
        return res
    
    def loss(self, X, Y):
        l = 0
        for (i, x) in enumerate(X):
            y_pred = self.feed_forward(x)
            l = l - np.log(y_pred @ self.to_1_hot(Y[i], isY = True))
        return l

    
    def feed_forward(self, X):
        V = X @ np.ones(2 * self.window)
        V = self.W_I.T @ V
        y = softmax(self.W_O.T @ V / (2 * self.window))
        return y
    
    def backprop(self, x, y, y_pred, gamma):
        V = self.W_I.T @ (x @ np.ones(2 * self.window)) / (2 * self.window)
        delta_W_O = np.outer(V, y_pred - y)
        
        V = self.W_O @ (y_pred - y) / (2 * self.window)
        x_ = x @ np.ones(2 * self.window)
        delta_W_I = np.outer(x_, V)
        
        self.W_O -= gamma * delta_W_O
        self.W_I -= gamma * delta_W_I
        
    def train(self, X, Y, gamma, epochs = 10):
        N = len(X)
        ind = np.arange(N)
        
        for e in range(epochs):
            np.random.shuffle(ind)
            X = X[ind]
            Y = Y[ind]
            for (i,x) in enumerate(X):
                x_ = self.to_1_hot(x)
                y_pred = self.feed_forward(x_)
                y = self.to_1_hot(Y[i], isY = True)
                self.backprop(x_, y, y_pred, gamma)
                
            print('loss at epoch {} is {}'.format(e, self.loss(X,Y)))
```

```python
model = embedder(window, V)
```

```python
model.train(X, Y, 0.01)
```
